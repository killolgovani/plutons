```r
#Making an object of the class Indices()

indices<-Indices()
```

```r
nifty_indices <- indices$NseTimeSeries() %>%
                    group_by(NAME) %>%
                    filter(!(NAME %like% '% TR%')) %>%
                    summarize(ST=min(TIME_STAMP), ET=max(TIME_STAMP)) %>%
                    print(n = Inf) 

%store nifty_indices
```
